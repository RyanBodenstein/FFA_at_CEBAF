```python
rng = rd.default_rng(secrets.randbits(128))

def cutnorm(sigma,num):
    dist = rng.normal(loc = 0.0, scale = sigma, size=num)
    for i in range(num):
        if dist[i] > 4*sigma:
            dist[i] = 2.5*sigma
        elif dist[i] < -4*sigma:
            dist[i] = -2.5*sigma
    return np.reshape(dist, (num,1))
```
